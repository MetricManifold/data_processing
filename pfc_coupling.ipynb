# PFC coupling

PFC fields can be coupled, and to implement this in SymPhas you have to know how to set the coefficients. In a PFC model, there are 5+1 different coefficients:
$$\alpha, \beta, \nu, \gamma, \delta$$
and there is also $\epsilon$.

These coefficients are indexed by both $i$ and $j$, and where $i=j$, they are the diagonal elements corresponding to the single field, all others are coupling between 
the indexed fields. So how is this organized in SymPhas?

The coefficients can be specified in a file. The organization is as follows:

- The coefficients for the individual fields (non coupling) appear first, and they are set in order of appearance (above) and in order of the fields.
- The coupling coefficients are next, and they appear in increasing order of fields and their coupled field. For example, in the case of four fields, they appear like this:
    - 1 and 2, 1 and 3, 1 and 4, 2 and 3, 2 and 4, 3 and 4.

Here is an example of how the coefficients file would look like for 4 fields:

```bash
[0..100]=0.00
[0..19]=1.      //alpha_{i, i}
[0..3]=-0.3
[4..7]=1
[8..11]=1
[12..15]=0      
[16..19]=1
[20..25]=-0.12  //alpha_{i, j}
[26..31]=0
[32..27]=0
[38..43]=0
[44..49]=0
[50..55]=1      //epsilon_{i, j}
```

So here, we've set all $\alpha_{i, j}$ where $i\ne j$ to $-0.12$.
